# Analisis Deskriptif
Dataset ini berisi Jumlah Kunjungan Wisatawan ke Obyek Wisata Unggulan Menurut Lokasi di Provinsi DKI Jakarta pada tahun 2008-2012

Penjelasan mengenai Variabel pada Dataset ini:

    tahun : Tahun
    (2008, 2009, 2010, 2011, 2012)
    
    lokasi : Lokasi wisata unggulan (Monumen Nasional, 
    Museum Nasional, Museum Satria Mandala, Museum 
    Sejarah Jakarta, Pelabuhan Sunda Kelapa, Ragunan, 
    TMII, Taman Impian Jaya Ancol)
    
    jumlah_kunjungan : Jumlah kunjungan wisatawan

Berdasarkan data di atas diketahui bahwa terdapat 40 data kunjungan selama 5 tahun berdasarkan 8 lokasi wisata unggulan. Provinsi DKI Jakarta memiliki jumlah kunjungan sebanyak 130,732,604 kunjungan pada tahun 2008-2012 dengan rata-rata kunjungan sebanyak 3,268,315.1 kunjungan.

Kunjungan terbanyak terjadi pada tahun 2012 sebanyak 30,067,363 kunjungan dan Taman Impian Jaya Ancol sebagai lokasi kunjungan terbanyak, yaitu sebanyak 73,622,225 kunjungan pada tahun 2008-2012. 

Kunjungan ke Taman Impian Jaya Ancol pada tahun 2012 merupakan kunjungan terbanyak, yaitu sebanyak 18,450,016 sedangkan kunjungan ke Monumen Nasional pada tahun 2008 merupakan kunjungan paling sedikit, yaitu sebanyak 12,677 kunjungan.

Berdasarkan diagram bar pertama terlihat bahwa tahun 2012 memiliki rata-rata kunjungan tertinggi, sedangkan tahun 2008 memiliki rata-rata kunjungan terendah dan terlihat bahwa Provinsi DKI Jakarta mengalami kenaikan jumlah kunjungan setiap tahunnya.

Berdasarkan diagram pie terlihat bahwa Taman Impian Jaya Ancol merupakan lokasi yang paling banyak dikunjungi, sedangkan Pelabuhan Sunda Kelapa merupakan lokasi yang paling sedikit dikunjungi pada tahun 2008-2012.

Berdasarkan diagram bar kedua terlihat bahwa Taman Impian jaya Ancol menjadi lokasi kunjungan terbanyak di setiap tahunnya sedangkan Pelabuhan Sunda Kelapa menjadi lokasi kunjungan paling sedikit di setiap tahunnya.

Oleh karena itu, data ini dapat dijadikan bahan evaluasi untuk pemerintah khususnya bagian pariwisata dan pengelola lokasi untuk lebih meningkatkan keamanan, kenyamanan, kesigapan dan segala sesuatunya di lokasi tersebut mengingat banyaknya wisatawan yang berkunjung serta melakukan rekonstruksi atau merenovasi lokasi yang minim kunjungan agar wisatawan lebih tertarik untuk berkunjung dan ikut andil dalam memajukan pariwisata Indonesia.
