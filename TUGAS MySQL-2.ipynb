```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# One to Many

```python
run("""
SELECT
    customerName,
    CONCAT(firstName," ", lastName) AS 'employeeName'
FROM
    customers
        INNER JOIN
    employees ON salesRepEmployeeNumber = employeeNumber
ORDER BY employeeName
""")
```

```python
run("""
SELECT 
    customerName,
    SUM(amount) AS totalPayment
FROM
    customers
        INNER JOIN
    payments USING (customerNumber)
WHERE
    customerName = 'Atelier graphique'
""")
```

```python
run("""
SELECT 
    paymentDate,
    SUM(amount) AS totalPayment
FROM
    payments
GROUP BY paymentDate
""")
```

```python
run("""
SELECT 
    productCode,
    productName,
    orderNumber
FROM
    products
        LEFT JOIN
    orderdetails USING (productCode)
WHERE
    orderNumber IS NULL
""")
```

```python
run("""
SELECT 
    customerName,
    customerNumber,
    SUM(amount) AS total
FROM
    customers
        INNER JOIN
    payments USING (customerNumber)
GROUP BY customerNumber
""")
```

```python
run("""
SELECT
    customerName,
    COUNT(*) AS jumlahOrder
FROM
    customers
        INNER JOIN
    orders USING (customerNumber)
WHERE customerName='Herkku Gifts'
""")
```

```python
run("""
SELECT
    CONCAT(firstName," ", lastName) AS 'employeeName',
    city
FROM
    employees
        INNER JOIN
    offices USING (officeCode)
WHERE city='Boston'
""")
```

```python
run("""
SELECT
    customerName,
    amount
FROM
    customers
        INNER JOIN
    payments USING (customerNumber)
WHERE amount>100000
ORDER BY amount DESC
""")
```

```python
run("""
SELECT
    SUM(quantityOrdered * priceEach) as nilaiOrder,
    status
FROM
    orders
        INNER JOIN
    orderdetails USING (orderNumber)
WHERE status='On Hold'
GROUP BY orderNumber
""")
```

```python
run("""
SELECT
    customerName,
    COUNT(*) AS statusOnHold
FROM
    customers
        INNER JOIN
    orders USING (customerNumber)
WHERE status='On Hold'
GROUP BY customerNumber
""")
```

# Many to Many Relationship

```python
run("""
SELECT
    orderDate,
    productName
FROM
    products 
        INNER JOIN orderdetails USING (productCode)
        INNER JOIN orders USING (orderNumber)
ORDER BY orderDate
""")
```

```python
run("""
SELECT
    productName,
    orderNumber,
    orderDate
FROM
    products 
        INNER JOIN orderdetails USING (productCode)
        INNER JOIN orders USING (orderNumber)
WHERE productName='1940 Ford Pickup Truck'
""")
```

```python
run("""
SELECT
    customerName,
    orderNumber,
    SUM(quantityOrdered * priceEach) as nilaiOrder
FROM
    customers 
        INNER JOIN orders USING (customerNumber)
        INNER JOIN orderdetails USING (orderNumber)
GROUP BY orderNumber
HAVING nilaiOrder>25000
""")
```

```python
run("""
SELECT
    productCode,
    COUNT(*) AS banyak
FROM
    orderdetails
GROUP BY productCode
HAVING banyak=(SELECT COUNT(*) FROM orders)
""")
```

```python
run("""
SELECT
    productName,
    priceEach,
    MSRP
FROM
    products
        INNER JOIN orderdetails USING (productCode)
WHERE priceEach<(0.8*MSRP)
ORDER BY priceEach
""")
```

```python
run("""
SELECT
    productName,
    priceEach-buyPrice AS Untung,
    buyPrice
FROM
    products
        INNER JOIN orderdetails USING (productCode)
WHERE priceEach-buyPrice>buyPrice
ORDER BY priceEach
""")
```

```python
run("""
SELECT
    dayName(orderDate),
    productName
FROM
    products
        INNER JOIN orderdetails USING (productCode)
        INNER JOIN orders USING (orderNumber)
WHERE DAYOFWEEK(orderdate)=2
""")
```

```python
run("""
SELECT
    status,
    SUM(quantityOrdered) as total
FROM
    orders 
        INNER JOIN orderdetails USING (orderNumber)
WHERE status='On Hold'
""")
```
